```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_raw = pd.read_csv('chilean.csv')
df = df_raw.copy()
print("datos importados")
```

```python
display(df.head(35))

print("\n---informacion de nuestro dataser----")
df.info()

print("\n---valores nulos por columna----")
print(df.isnull().sum())
```

```python
#para quitar los separadores en los titulos de las columnas

df.columns = [col.strip().replace(' ', '_') for col in df.columns]

display(df.head())
```

```python
#Columna 'recipe_name'
print('Valores nulos es la columna " recipe_name": ', df['recipe_name'].isnull().sum())

df['recipe_name'] = df['recipe_name'].str.capitalize()

df=df.dropna(subset=['recipe_name']) 

display(df.head(10))
```

```python
#Columna 'preparation_time'

print('Valores nulos es la columna "preparation_time": ', df['preparation_time'].isnull().sum())

df['preparation_time'] = pd.to_numeric(df['preparation_time'], errors = 'coerce')

mediana_preptime = df['preparation_time'].median()
df['preparation_time'] = df['preparation_time'].fillna(mediana_preptime, inplace=False)

media_sin_ceros = df[df['preparation_time'] != 0]['preparation_time'].median()

df['preparation_time'] = df['preparation_time'].replace(0, media_sin_ceros)

df['preparation_time'] = df['preparation_time'].astype(int)

display(df.head(10))



```

```python
#columna 'min_portion'
print('Valores nulos es la columna "min_portion": ', df['min_portion'].isnull().sum())
df['min_portion'] = pd.to_numeric(df['min_portion'], errors = 'coerce')

mediana_min_portion = df['min_portion'].median()
df['min_portion'] = df['min_portion'].fillna(mediana_min_portion, inplace=False)

df['min_portion'] = df['min_portion'].astype(int)

display(df.head(10))

```

```python

#columna 'max_portion'
print('Valores nulos es la columna "max_portion": ', df['max_portion'].isnull().sum())

df['max_portion'] = pd.to_numeric(df['max_portion'], errors = 'coerce')


df['max_portion'] = df['max_portion'].fillna(df['min_portion'])

df['max_portion'] = df['max_portion'].astype(int)

display(df.head(10))
```

```python
#columna 'ingridients'
print('Valores nulos es la columna "ingridients": ', df['ingredients'].isnull().sum())

df = df.dropna(subset=['ingredients'])

df['ingredients'] = df['ingredients'].str.capitalize()

df['ingredients'] = df['ingredients'].str.replace(r"[\[\]\"']", '', regex=True)

display(df.head(10))
```

```python
#columna 'steps'
print('Valores nulos es la columna "steps": ', df['steps'].isnull().sum())

df = df.dropna(subset=['steps'])

df['steps'] = df['steps'].str.capitalize()

df['steps'] = df['steps'].str.replace(r"[\[\]\"']", '', regex=True)

display(df.head(10))
```

```python
#new column ingredient count
#gracias a apply recorreremos la columna ingredientes y junto a la funcion lambda 
#contara las comas y las determinara como un ingrediente
df['ingredient_count'] = df['ingredients'].apply(lambda x: len(str(x).split(',')) if pd.notnull(x) else 0)

display(df.head(10))
```

```python
#new column nivel de dificultad por tiempo

def categorizar_tiempo(tiempo):
    if tiempo <= 20:
        return 'duracion Rapida'
    elif tiempo <= 60:
        return 'duracion Media '
    else:
        return 'duracion Larga'

df['duration_level'] = df['preparation_time'].apply(categorizar_tiempo)

display(df.head(10))
```

```python
#new column, nivel de complejidad de receta

max_time = df['preparation_time'].max()
max_ing = df['ingredient_count'].max()

df['complexity_score'] = ((df['preparation_time'] / max_time) * 50) + ((df['ingredient_count'] / max_ing) * 50)

display(df.head(10))
```

```python
#exportar
df.to_csv('datos_limpios.csv', index=False,encoding='utf-8')
df.to_excel('datos_limpios.xlsx')
```
